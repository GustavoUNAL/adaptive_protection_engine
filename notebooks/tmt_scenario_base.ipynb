# Anlisis detallado escenario base

```python
# %% Import Libraries
import json
import os
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
import matplotlib.pyplot as plt  # Nueva importación para histogramas
import traceback

# %% Configuration

# --- File Paths ---
# !! IMPORTANTE: Actualiza estas rutas !!
json_file_before = "/Users/gustavo/Documents/Projects/TESIS_UNAL/ADAPTIVE_ALGORITHM/data/processed/independent_relay_pairs_automation.json"  # Datos ANTES de optimización
json_file_after = "/Users/gustavo/Documents/Projects/TESIS_UNAL/ADAPTIVE_ALGORITHM/data/processed/independent_relay_pairs_optimization.json"   # Datos DESPUÉS de optimización

# --- Scenario to Analyze ---
TARGET_SCENARIO_ID = "scenario_1"  # Escenario a comparar

# --- Output Directory ---
output_dir_comparison = "/Users/gustavo/Documents/Projects/TESIS_UNAL/ADAPTIVE_ALGORITHM/results/comparison_plots_thesis_detailed_interactive_v4"  # Directorio para resultados

# --- Parameters ---
CTI = 0.2  # Intervalo de Tiempo de Coordinación (s)
FLOAT_COMPARISON_TOLERANCE = 1e-6  # Tolerancia para comparar ajustes flotantes
DPI = 300  # Resolución para imágenes (alta calidad para tesis)

# --- Plotting Style (Thesis Quality - Español - COLORES AZUL/ROJO) ---
pio.templates.default = "plotly_white"
FIG_WIDTH_THESIS_BAR = 1100
FIG_HEIGHT_THESIS_BAR = 650
FIG_WIDTH_SUBPLOTS = 1200
FIG_HEIGHT_SUBPLOTS = 900
FIG_WIDTH_HIST = 9  # En pulgadas para Matplotlib
FIG_HEIGHT_HIST = 6  # En pulgadas para Matplotlib
TITLE_FONT_SIZE_THESIS = 18
AXIS_LABEL_FONT_SIZE_THESIS = 14
TICK_FONT_SIZE_THESIS = 12
LEGEND_TITLE_FONT_SIZE_THESIS = 14
LEGEND_FONT_SIZE_THESIS = 12
SUBPLOT_TITLE_FONT_SIZE = 15
ANNOTATION_FONT_SIZE_THESIS = 11
TEMPLATE_THESIS = "plotly_white"
COLOR_BEFORE = '#1f77b4'  # Azul Plotly (Antes Opt.) - Formato hexadecimal para compatibilidad
COLOR_AFTER = '#d62728'   # Rojo Plotly (Después Opt.)
HOVER_BG_COLOR = 'rgba(245, 245, 245, 0.95)'

# %% Helper Functions (Sin cambios)

def calculate_MT(tm, tb, cti=CTI):
    if not isinstance(tm, (int, float)) or not isinstance(tb, (int, float)):
        return None, None
    delta_tmb = tb - tm - cti
    MT = (delta_tmb - abs(delta_tmb)) / 2
    return MT, delta_tmb

def load_and_analyze_scenario(filepath, target_scenario_id, cti=CTI):
    print(f"--- Procesando archivo: {os.path.basename(filepath)} para Escenario: {target_scenario_id} ---")
    if not os.path.exists(filepath):
        print(f"Error: Archivo no encontrado: {filepath}")
        return None
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
        if not isinstance(data, list):
            raise TypeError(f"Se esperaba una Lista JSON '[]', se encontró tipo '{type(data)}'.")
        if not data:
            print("Advertencia: Archivo JSON vacío.")
            return pd.DataFrame()
    except Exception as e:
        print(f"Error cargando JSON {filepath}: {e}")
        return None

    scenario_pairs_data = []
    processed_count = 0
    skipped_count = 0
    found_scenario = False
    missing_settings_warn = False
    settings_map = {'TDS': 'TDS', 'pick_up': 'Pickup'}

    for idx, relay_pair in enumerate(data):
        if not isinstance(relay_pair, dict):
            skipped_count += 1
            continue
        scenario_id = relay_pair.get("scenario_id")
        if scenario_id == target_scenario_id:
            found_scenario = True
            main_relay = relay_pair.get("main_relay")
            backup_relay = relay_pair.get("backup_relay")
            if not isinstance(main_relay, dict) or not isinstance(backup_relay, dict):
                skipped_count += 1
                continue
            tm_val = main_relay.get("Time_out")
            tb_val = backup_relay.get("Time_out")
            main_relay_id = main_relay.get("relay", f"M_Unk_{idx+1}")
            backup_relay_id = backup_relay.get("relay", f"B_Unk_{idx+1}")
            pair_name = f"{main_relay_id}-{backup_relay_id}"
            if tm_val is None or tb_val is None:
                skipped_count += 1
                continue
            try:
                tm_float = float(tm_val)
                tb_float = float(tb_val)
                if tm_float < 0 or tb_float < 0:
                    skipped_count += 1
                    continue
                mt, delta_t = calculate_MT(tm_float, tb_float, cti)
                if mt is None:
                    skipped_count += 1
                    continue
            except (ValueError, TypeError):
                skipped_count += 1
                continue

            main_settings = {}
            backup_settings = {}
            settings_found_main = True
            settings_found_backup = True
            for source_key, target_key in settings_map.items():
                try:
                    main_settings[f'main_{target_key}'] = float(main_relay.get(source_key))
                except (TypeError, ValueError, AttributeError):
                    main_settings[f'main_{target_key}'] = np.nan
                    settings_found_main = False
                try:
                    backup_settings[f'backup_{target_key}'] = float(backup_relay.get(source_key))
                except (TypeError, ValueError, AttributeError):
                    backup_settings[f'backup_{target_key}'] = np.nan
                    settings_found_backup = False

            if not settings_found_main or not settings_found_backup:
                if not missing_settings_warn:
                    print(f"  Advertencia: Ajustes TDS/pick_up faltantes en {os.path.basename(filepath)}. Serán NaN.")
                    missing_settings_warn = True

            pair_info = {
                "pair_label": pair_name,
                "mt": mt,
                "delta_t": delta_t,
                "tm": tm_float,
                "tb": tb_float,
                "main_relay_id": main_relay_id,
                "backup_relay_id": backup_relay_id,
                "line": relay_pair.get('line', main_relay.get('line', 'N/A')),
                "fault": relay_pair.get('fault', main_relay.get('fault', 'N/A')),
                "status": "Coordinado" if delta_t >= 0 else "Descoordinado",
                **main_settings,
                **backup_settings
            }
            scenario_pairs_data.append(pair_info)
            processed_count += 1

    if not found_scenario:
        print(f"Advertencia: Escenario '{target_scenario_id}' no encontrado.")
        return pd.DataFrame()
    print(f"  {processed_count} pares válidos extraídos, {skipped_count} registros omitidos.")
    if not scenario_pairs_data:
        print(f"  No se extrajeron pares válidos para {target_scenario_id}.")
        return pd.DataFrame()

    all_cols = ["pair_label", "mt", "delta_t", "tm", "tb", "main_relay_id", "backup_relay_id",
                "line", "fault", "status", "main_TDS", "main_Pickup", "backup_TDS", "backup_Pickup"]
    df = pd.DataFrame(scenario_pairs_data)
    for col in all_cols:
        if col not in df.columns:
            df[col] = np.nan
    return df[all_cols]

# %% Plotting Function for Pair-wise MT Comparison (Sin cambios)
def plot_mt_comparison_with_settings_thesis_es(df_merged_all, scenario_id, output_dir, show_plots=True):
    print(f"\n--- Generando Gráfico Comparativo MT por Par para Escenario: {scenario_id} ---")
    df_plot = df_merged_all.dropna(subset=['mt_before', 'mt_after']).copy()
    setting_cols = [col for col in df_plot.columns if 'TDS' in col or 'Pickup' in col]
    df_plot[setting_cols] = df_plot[setting_cols].fillna('N/A')
    if df_plot.empty:
        print(f"  No se encontraron pares comunes. Omitiendo gráfico MT.")
        return
    df_plot.sort_values(by=['mt_before', 'pair_label'], ascending=[True, True], inplace=True)
    fig = go.Figure()
    hover_template = """<b>Par:</b> %{x}<br>--------------------------------<br><b>MT Antes:</b> %{customdata[0]:.4f} | <b>MT Después:</b> %{customdata[1]:.4f} (<b>Cambio:</b> %{customdata[2]:+.4f})<br><b>Δt Antes:</b> %{customdata[3]:.4f}s | <b>Δt Después:</b> %{customdata[4]:.4f}s<br><b>Estado Antes:</b> %{customdata[5]} | <b>Estado Después:</b> %{customdata[6]}<br>--------------------------------<br><u>Ajustes Antes:</u><br>  Principal: TDS=%{customdata[7]}, Pickup=%{customdata[8]}<br>  Respaldo: TDS=%{customdata[9]}, Pickup=%{customdata[10]}<br><u>Ajustes Después:</u><br>  Principal: TDS=%{customdata[11]}, Pickup=%{customdata[12]}<br>  Respaldo: TDS=%{customdata[13]}, Pickup=%{customdata[14]}<br><extra></extra>"""
    custom_data_cols = ['mt_before', 'mt_after', 'mt_improvement', 'delta_t_before', 'delta_t_after',
                       'status_before', 'status_after', 'main_TDS_before', 'main_Pickup_before',
                       'backup_TDS_before', 'backup_Pickup_before', 'main_TDS_after', 'main_Pickup_after',
                       'backup_TDS_after', 'backup_Pickup_after']
    for col in custom_data_cols:
        if col not in df_plot.columns:
            df_plot[col] = 'N/A'
    customdata_array = df_plot[custom_data_cols].values
    fig.add_trace(go.Bar(
        x=df_plot['pair_label'], y=df_plot['mt_before'], name='MT Antes Opt.',
        marker_color=COLOR_BEFORE, offsetgroup=1, customdata=customdata_array, hovertemplate=hover_template
    ))
    fig.add_trace(go.Bar(
        x=df_plot['pair_label'], y=df_plot['mt_after'], name='MT Después Opt.',
        marker_color=COLOR_AFTER, offsetgroup=2, customdata=customdata_array, hovertemplate=hover_template
    ))
    fig.update_layout(
        template=TEMPLATE_THESIS, width=FIG_WIDTH_THESIS_BAR, height=FIG_HEIGHT_THESIS_BAR,
        title=f'Comparación Índice de Descoordinación (MT) por Par: Antes vs Después<br>Escenario: {scenario_id}',
        xaxis_title='Par de Relés', yaxis_title='Índice de Descoordinación (MT)',
        font=dict(family="Arial, sans-serif", size=TICK_FONT_SIZE_THESIS),
        title_font_size=TITLE_FONT_SIZE_THESIS, xaxis_title_font_size=AXIS_LABEL_FONT_SIZE_THESIS,
        yaxis_title_font_size=AXIS_LABEL_FONT_SIZE_THESIS, xaxis_tickfont_size=TICK_FONT_SIZE_THESIS,
        yaxis_tickfont_size=TICK_FONT_SIZE_THESIS,
        legend=dict(title='Estado', title_font_size=LEGEND_TITLE_FONT_SIZE_THESIS, font_size=LEGEND_FONT_SIZE_THESIS,
                    yanchor="top", y=0.98, xanchor="left", x=0.01, bgcolor='rgba(255,255,255,0.7)',
                    bordercolor="LightGrey", borderwidth=1),
        barmode='group', bargap=0.2, bargroupgap=0.1, margin=dict(l=70, r=40, t=100, b=180),
        hoverlabel=dict(bgcolor=HOVER_BG_COLOR, bordercolor="darkgrey", font_size=11, align='left'),
        xaxis=dict(gridcolor='LightGray', linecolor='black', tickangle=-75),
        yaxis=dict(gridcolor='LightGray', linecolor='black', zeroline=True, zerolinecolor='black', zerolinewidth=1.5)
    )
    fig.add_hline(y=0, line_width=2, line_dash="dash", line_color="black",
                  annotation_text="MT = 0 (Umbral Coord.)", annotation_position="bottom right",
                  annotation_font_size=ANNOTATION_FONT_SIZE_THESIS)
    if output_dir:
        print(f"  Guardando gráfico comparativo MT por par...")
        safe_scenario_id = scenario_id.replace(" ", "_").replace("/", "-")
        base_filename = os.path.join(output_dir, f"Escenario_{safe_scenario_id}_Comparacion_MT_Pares_Tesis")
        try:
            fig.write_html(f"{base_filename}.html")
            print(f"    Guardado: {base_filename}.html")
        except Exception as e:
            print(f"    Error guardando HTML '{base_filename}.html': {e}")
        for fmt in ['svg', 'pdf', 'png']:
            static_filename = f"{base_filename}.{fmt}"
            try:
                pio.write_image(fig, static_filename, format=fmt, scale=2 if fmt == 'png' else 1)
                print(f"    Guardado: {static_filename}")
            except Exception as e:
                print(f"    Error guardando imagen estática '{static_filename}': {e}")
    if show_plots:
        print(f"    Mostrando gráfico comparativo MT por par...")
        fig.show()

# %% Plotting Function for COMBINED Settings Comparison (Sin cambios)
def plot_all_settings_comparison_bars_thesis_es(df_merged_all, scenario_id, output_dir, show_plots=True):
    print(f"\n--- Generando Gráfico Combinado Comparativo de AJUSTES para Escenario: {scenario_id} ---")
    settings_info = {
        'main_TDS': {'label': 'TDS Relé Principal', 'row': 1, 'col': 1},
        'main_Pickup': {'label': 'Pickup Relé Principal', 'row': 1, 'col': 2},
        'backup_TDS': {'label': 'TDS Relé Respaldo', 'row': 2, 'col': 1},
        'backup_Pickup': {'label': 'Pickup Relé Respaldo', 'row': 2, 'col': 2}
    }
    setting_cols_before = [f"{s}_before" for s in settings_info.keys()]
    setting_cols_after = [f"{s}_after" for s in settings_info.keys()]
    df_plot = df_merged_all.dropna(subset=setting_cols_before + setting_cols_after).copy()
    if df_plot.empty:
        print(f"  No se encontraron pares con TODOS los ajustes válidos. Omitiendo gráfico combinado.")
        return
    df_plot.sort_values(by='pair_label', ascending=True, inplace=True)
    print(f"  Generando gráfico combinado para {len(df_plot)} pares con ajustes completos.")
    subplot_titles = [info['label'] for info in settings_info.values()]
    fig = make_subplots(rows=2, cols=2, subplot_titles=subplot_titles)
    show_legend_flag = True
    for setting_prefix, info in settings_info.items():
        setting_before_col = f'{setting_prefix}_before'
        setting_after_col = f'{setting_prefix}_after'
        axis_label = info['label']
        row, col = info['row'], info['col']
        df_plot[f'{setting_prefix}_change'] = df_plot[setting_after_col] - df_plot[setting_before_col]
        hover_template_setting = """<b>Par:</b> %{x}<br>--------------------------------<br><b>%{customdata[2]}:</b><br>  Antes: %{customdata[0]}<br>  Después: %{customdata[1]}<br>  <b>Cambio: %{customdata[3]:+.4f}</b><br>--------------------------------<br>MT Antes: %{customdata[4]:.4f} | MT Después: %{customdata[5]:.4f}<br><extra></extra>"""
        custom_data_cols_setting = [setting_before_col, setting_after_col, 'setting_label',
                                   f'{setting_prefix}_change', 'mt_before', 'mt_after']
        df_plot['setting_label'] = axis_label
        if 'mt_before' not in df_plot.columns:
            df_plot['mt_before'] = np.nan
        if 'mt_after' not in df_plot.columns:
            df_plot['mt_after'] = np.nan
        customdata_array_setting = df_plot[custom_data_cols_setting].values
        fig.add_trace(
            go.Bar(x=df_plot['pair_label'], y=df_plot[setting_before_col], name='Antes Opt.',
                   marker_color=COLOR_BEFORE, offsetgroup=f'group{row}{col}', legendgroup='antes',
                   showlegend=show_legend_flag, customdata=customdata_array_setting,
                   hovertemplate=hover_template_setting),
            row=row, col=col
        )
        fig.add_trace(
            go.Bar(x=df_plot['pair_label'], y=df_plot[setting_after_col], name='Después Opt.',
                   marker_color=COLOR_AFTER, offsetgroup=f'group{row}{col}+1', legendgroup='despues',
                   showlegend=show_legend_flag, customdata=customdata_array_setting,
                   hovertemplate=hover_template_setting),
            row=row, col=col
        )
        show_legend_flag = False
    fig.update_layout(
        template=TEMPLATE_THESIS, width=FIG_WIDTH_SUBPLOTS, height=FIG_HEIGHT_SUBPLOTS,
        title=f'Comparación de Ajustes (TDS/Pickup) por Par: Antes vs Después<br>Escenario: {scenario_id}',
        font=dict(family="Arial, sans-serif", size=TICK_FONT_SIZE_THESIS),
        title_font_size=TITLE_FONT_SIZE_THESIS,
        legend=dict(title='Estado', title_font_size=LEGEND_TITLE_FONT_SIZE_THESIS, font_size=LEGEND_FONT_SIZE_THESIS,
                    orientation="h", yanchor="bottom", y=1.01, xanchor="center", x=0.5),
        barmode='group', bargap=0.2, bargroupgap=0.1,
        hoverlabel=dict(bgcolor=HOVER_BG_COLOR, bordercolor="darkgrey", font_size=11, align='left'),
        margin=dict(l=70, r=50, t=120, b=180)
    )
    for setting_prefix, info in settings_info.items():
        axis_label = info['label']
        row, col = info['row'], info['col']
        fig.update_xaxes(title_text="Par de Relés", row=row, col=col, tickangle=-75,
                         title_font_size=AXIS_LABEL_FONT_SIZE_THESIS - 2, tickfont_size=TICK_FONT_SIZE_THESIS - 1)
        fig.update_yaxes(title_text=f"Valor {axis_label}", row=row, col=col,
                         title_font_size=AXIS_LABEL_FONT_SIZE_THESIS - 2, tickfont_size=TICK_FONT_SIZE_THESIS)
    for annotation in fig.layout.annotations:
        annotation.font.size = SUBPLOT_TITLE_FONT_SIZE
    if output_dir:
        print(f"  Guardando gráfico combinado de comparación de ajustes...")
        safe_scenario_id = scenario_id.replace(" ", "_").replace("/", "-")
        base_filename = os.path.join(output_dir, f"Escenario_{safe_scenario_id}_Comparacion_Ajustes_Combinado_Tesis")
        try:
            fig.write_html(f"{base_filename}.html")
            print(f"    Guardado: {base_filename}.html")
        except Exception as e:
            print(f"    Error guardando HTML '{base_filename}.html': {e}")
        for fmt in ['svg', 'pdf', 'png']:
            static_filename = f"{base_filename}.{fmt}"
            try:
                pio.write_image(fig, static_filename, format=fmt, scale=2 if fmt == 'png' else 1)
                print(f"    Guardado: {static_filename}")
            except Exception as e:
                print(f"    Error guardando imagen estática '{static_filename}': {e}")
    if show_plots:
        print(f"    Mostrando gráfico combinado de comparación de ajustes...")
        fig.show()

# %% Plotting Function for Distribution Comparison (NUEVO: Usando Matplotlib)
def plot_distribution_comparison_bars_thesis_es(df_before, df_after, scenario_id, output_dir, metric='delta_t', show_plots=True):
    """
    Genera histogramas agrupados normalizados usando Matplotlib para comparar una métrica
    (MT o Delta_t) antes y después de la optimización. Guarda imágenes en alta resolución.
    """
    print(f"\n--- Generando Histograma AGRUPADO y NORMALIZADO {metric.upper()} para Escenario: {scenario_id} ---")
    has_before = not df_before.empty and metric in df_before.columns
    has_after = not df_after.empty and metric in df_after.columns
    if not has_before and not has_after:
        print(f"  Sin datos para {metric}. Omitiendo gráfico.")
        return

    # Determinar rango y bins
    min_val, max_val = np.inf, -np.inf
    if has_before:
        min_val = min(min_val, df_before[metric].min())
        max_val = max(max_val, df_before[metric].max())
    if has_after:
        min_val = min(min_val, df_after[metric].min())
        max_val = max(max_val, df_after[metric].max())
    if min_val == np.inf or max_val == -np.inf or min_val >= max_val:
        print(f"  Rango inválido para {metric}. Omitiendo gráfico.")
        return
    if min_val == max_val:
        min_val -= 0.01
        max_val += 0.01

    num_bins = 25
    bin_size = (max_val - min_val) / num_bins
    if bin_size <= 1e-9:
        bin_size = (abs(max_val) + abs(min_val) + 1e-6) / num_bins if max_val != 0 or min_val != 0 else 1e-6 / num_bins
    bin_edges = np.linspace(min_val, max_val, num_bins + 1)

    # Configurar figura
    plt.figure(figsize=(FIG_WIDTH_HIST, FIG_HEIGHT_HIST), dpi=DPI)
    plt.style.use('seaborn-whitegrid')  # Estilo limpio para publicaciones

    # Calcular histogramas
    hist_before = np.histogram(df_before[metric], bins=bin_edges, density=True)[0] if has_before else np.zeros(num_bins)
    hist_after = np.histogram(df_after[metric], bins=bin_edges, density=True)[0] if has_after else np.zeros(num_bins)

    # Configurar barras agrupadas
    bar_width = bin_size / 2.2  # Ancho de cada barra (ligeramente reducido para claridad)
    bin_centers = bin_edges[:-1] + bin_size / 2
    offset = bar_width / 2

    if has_before:
        plt.bar(bin_centers - offset, hist_before, width=bar_width, color=COLOR_BEFORE, alpha=0.85,
                label='Antes Opt.', edgecolor='white')
    if has_after:
        plt.bar(bin_centers + offset, hist_after, width=bar_width, color=COLOR_AFTER, alpha=0.85,
                label='Después Opt.', edgecolor='white')

    # Configurar ejes y título
    axis_title_es = f'Diferencia de Tiempo Δt = T_b - T_m - CTI [s]' if metric == 'delta_t' else 'Índice de Descoordinación (MT)'
    plot_title_es = f'Distribución de {axis_title_es.split("[")[0]} (Barras Agrupadas)\nEscenario: {scenario_id}'
    plt.title(plot_title_es, fontsize=TITLE_FONT_SIZE_THESIS, pad=15, family='Arial')
    plt.xlabel(axis_title_es, fontsize=AXIS_LABEL_FONT_SIZE_THESIS, family='Arial')
    plt.ylabel('Densidad de Probabilidad', fontsize=AXIS_LABEL_FONT_SIZE_THESIS, family='Arial')
    plt.tick_params(axis='both', labelsize=TICK_FONT_SIZE_THESIS)

    # Añadir línea en x=0
    plt.axvline(x=0, color='black', linestyle='--', linewidth=1.5)
    if metric == 'delta_t':
        plt.text(0.98, 0.02, 'Umbral Coord. (Δt=0)', transform=plt.gca().transAxes,
                 fontsize=ANNOTATION_FONT_SIZE_THESIS, ha='right', va='bottom',
                 bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

    # Configurar leyenda
    plt.legend(title='Estado', fontsize=LEGEND_FONT_SIZE_THESIS, title_fontsize=LEGEND_TITLE_FONT_SIZE_THESIS,
               loc='upper right', frameon=True, edgecolor='lightgrey')

    # Ajustar diseño
    plt.tight_layout()

    # Guardar figura
    if output_dir:
        print(f"  Guardando histograma agrupado y normalizado de {metric.upper()}...")
        safe_scenario_id = scenario_id.replace(" ", "_").replace("/", "-")
        base_filename = os.path.join(output_dir, f"Escenario_{safe_scenario_id}_Distribucion_{metric.upper()}_Agrupada_Normalizada_Tesis")
        for fmt in ['png', 'svg', 'pdf']:
            try:
                plt.savefig(f"{base_filename}.{fmt}", format=fmt, dpi=DPI, bbox_inches='tight', transparent=True)
                print(f"    Guardado: {base_filename}.{fmt}")
            except Exception as e:
                print(f"    Error guardando '{base_filename}.{fmt}': {e}")
    else:
        print("  Advertencia: Directorio de salida no especificado, no se guardará histograma.")

    # Mostrar gráfico
    if show_plots:
        print(f"    Mostrando histograma agrupado y normalizado de {metric.upper()}...")
        plt.show()
    plt.close()

# %% Main Execution Logic
def main_detailed_comparison_thesis_spanish_interactive():
    print("===================================================================================")
    print("=== Comparación Detallada Antes/Después Optimización (Tesis, Español, Interactivo) ===")
    print(f"=== Escenario: {TARGET_SCENARIO_ID} | CTI: {CTI}s ===")
    print("===================================================================================")

    try:
        os.makedirs(output_dir_comparison, exist_ok=True)
        print(f"Resultados y gráficos se guardarán en: {output_dir_comparison}")
    except OSError as e:
        print(f"Error creando directorio de salida: {e}")
        return

    df_before = load_and_analyze_scenario(json_file_before, TARGET_SCENARIO_ID, CTI)
    df_after = load_and_analyze_scenario(json_file_after, TARGET_SCENARIO_ID, CTI)

    if df_before is None or df_after is None:
        print("\nError Crítico: No se pudieron cargar los datos. Saliendo.")
        return
    if df_before.empty and df_after.empty:
        print(f"\nComparación abortada: No hay datos para el escenario '{TARGET_SCENARIO_ID}'.")
        return

    expected_cols = ['pair_label', 'mt', 'delta_t', 'tm', 'tb', 'status', 'main_TDS', 'main_Pickup', 'backup_TDS', 'backup_Pickup']
    for df in [df_before, df_after]:
        if not df.empty:
            for col in expected_cols:
                if col not in df.columns:
                    df[col] = np.nan
    df_merged = pd.merge(df_before, df_after, on='pair_label', how='outer', suffixes=('_antes', '_despues'))
    rename_map = {
        'mt_antes': 'mt_before', 'mt_despues': 'mt_after',
        'delta_t_antes': 'delta_t_before', 'delta_t_despues': 'delta_t_after',
        'status_antes': 'status_before', 'status_despues': 'status_after',
        'main_TDS_antes': 'main_TDS_before', 'main_TDS_despues': 'main_TDS_after',
        'main_Pickup_antes': 'main_Pickup_before', 'main_Pickup_despues': 'main_Pickup_after',
        'backup_TDS_antes': 'backup_TDS_before', 'backup_TDS_despues': 'backup_TDS_after',
        'backup_Pickup_antes': 'backup_Pickup_before', 'backup_Pickup_despues': 'backup_Pickup_after',
    }
    df_merged.rename(columns=rename_map, inplace=True)
    df_merged['mt_improvement'] = df_merged['mt_after'] - df_merged['mt_before']
    df_merged['delta_t_improvement'] = df_merged['delta_t_after'] - df_merged['delta_t_before']
    df_merged['status_before'] = df_merged['status_before'].fillna('No Encontrado')
    df_merged['status_after'] = df_merged['status_after'].fillna('No Encontrado')
    df_merged['status_before'].replace({'Coordinated': 'Coordinado', 'Miscoordinated': 'Descoordinado'}, inplace=True)
    df_merged['status_after'].replace({'Coordinated': 'Coordinado', 'Miscoordinated': 'Descoordinado'}, inplace=True)

    settings_to_check = ['main_TDS', 'main_Pickup', 'backup_TDS', 'backup_Pickup']
    setting_change_counts = {}
    for setting in settings_to_check:
        col_before = f'{setting}_before'
        col_after = f'{setting}_after'
        if col_before in df_merged.columns and col_after in df_merged.columns:
            valid_mask = df_merged[col_before].notna() & df_merged[col_after].notna()
            df_merged[f'{setting}_changed'] = False
            df_merged.loc[valid_mask, f'{setting}_changed'] = ~np.isclose(
                df_merged.loc[valid_mask, col_after], df_merged.loc[valid_mask, col_before],
                rtol=FLOAT_COMPARISON_TOLERANCE, atol=FLOAT_COMPARISON_TOLERANCE, equal_nan=False
            )
            setting_change_counts[setting] = df_merged[f'{setting}_changed'].sum()
        else:
            setting_change_counts[setting] = 0

    print(f"\n--- Resumen Comparativo para Escenario: {TARGET_SCENARIO_ID} ---")
    valid_comparison_mask = (df_merged['status_before'] != 'No Encontrado') & (df_merged['status_after'] != 'No Encontrado')
    n_compared = valid_comparison_mask.sum()
    if n_compared > 0:
        tmt_before = df_merged.loc[valid_comparison_mask, 'mt_before'].sum()
        tmt_after = df_merged.loc[valid_comparison_mask, 'mt_after'].sum()
        miscoord_before = (df_merged.loc[valid_comparison_mask, 'status_before'] == 'Descoordinado').sum()
        miscoord_after = (df_merged.loc[valid_comparison_mask, 'status_after'] == 'Descoordinado').sum()
        print(f"Total Pares Comparados (presentes en ambos): {n_compared}")
        print("\nMétricas de Coordinación:")
        print(f"  TMT Antes: {tmt_before:10.4f}   |   TMT Después: {tmt_after:10.4f}   |   Cambio: {tmt_after - tmt_before:+10.4f}")
        print(f"  Pares Desc. (#) Antes: {miscoord_before:<3} |   Pares Desc. (#) Después: {miscoord_after:<3} |   Cambio: {miscoord_after - miscoord_before:+4}")
        pct_mis_before = (miscoord_before / n_compared * 100) if n_compared > 0 else 0
        pct_mis_after = (miscoord_after / n_compared * 100) if n_compared > 0 else 0
        print(f"  Pares Desc. (%) Antes: {pct_mis_before:5.1f}% |   Pares Desc. (%) Después: {pct_mis_after:5.1f}% |   Cambio: {pct_mis_after - pct_mis_before:+6.1f}%p")
        print("\nCambios en Ajustes (Pares con modificación):")
        for setting, count in setting_change_counts.items():
            pct_changed = (count / n_compared * 100) if n_compared > 0 else 0
            setting_label = setting.replace('_', ' ').replace('main', 'Principal').replace('backup', 'Respaldo').replace('Pickup', 'Pickup')
            print(f"  - {setting_label:<18}: {count:>5,} pares cambiaron ({pct_changed:5.1f}%)")
    else:
        print("No se encontraron pares comunes entre los datasets 'antes' y 'después'.")

    SHOW_PLOTS_INTERACTIVELY = True

    if not df_merged.empty:
        if n_compared > 0:
            plot_mt_comparison_with_settings_thesis_es(
                df_merged, TARGET_SCENARIO_ID, output_dir_comparison, show_plots=SHOW_PLOTS_INTERACTIVELY
            )
        plot_all_settings_comparison_bars_thesis_es(
            df_merged, TARGET_SCENARIO_ID, output_dir_comparison, show_plots=SHOW_PLOTS_INTERACTIVELY
        )
        plot_distribution_comparison_bars_thesis_es(
            df_before if df_before is not None else pd.DataFrame(),
            df_after if df_after is not None else pd.DataFrame(),
            TARGET_SCENARIO_ID, output_dir_comparison, metric='delta_t', show_plots=SHOW_PLOTS_INTERACTIVELY
        )
        plot_distribution_comparison_bars_thesis_es(
            df_before if df_before is not None else pd.DataFrame(),
            df_after if df_after is not None else pd.DataFrame(),
            TARGET_SCENARIO_ID, output_dir_comparison, metric='mt', show_plots=SHOW_PLOTS_INTERACTIVELY
        )
    else:
        print("\nOmitiendo generación de gráficos.")

    print("\nGeneración y visualización de gráficos comparativos detallados completa.")

# %% Script Entry Point
if __name__ == "__main__":
    try:
        main_detailed_comparison_thesis_spanish_interactive()
    except FileNotFoundError as e:
        print(f"\n--- Error: Archivo No Encontrado ---\n{e}\n*** Asegúrate que las rutas a los archivos JSON sean correctas. ***")
    except Exception as e:
        print(f"\n--- Ocurrió un Error Inesperado ---\nTipo Error: {type(e).__name__}\nDetalles: {str(e)}\n--- Traceback ---\n{traceback.format_exc()}-----------------")

    print("\n===================================================================================")
    print("=== Ejecución del Script de Comparación Detallada Finalizada ===")
    print("===================================================================================")
```
