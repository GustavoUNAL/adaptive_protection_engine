```python
import pandas as pd
import numpy as np
import json

# 1. Cargar datos
df = pd.read_csv(
    '/Users/gustavo/Documents/Projects/TESIS_UNAL/ADAPTIVE_ALGORITHM/scripts/all_scenarios/summary_scenarios_optimized.csv'
)

# 2. Renombrar columnas
df.rename(columns={
    'TMT (negativo)': 'TMT_negativo',
    '% Descoord': 'pct_descoord'
}, inplace=True)

# 3. Ordenar y calcular umbral
df_sorted = df.sort_values('TMT_negativo')
threshold = df_sorted['TMT_negativo'].median()

# 4. Clasificar índices
df_sorted['indice'] = np.where(
    df_sorted['TMT_negativo'] >= threshold,
    'Bueno',
    'Malo'
)

# 5. Filtrar solo los “Buenos” y extraer sus IDs
buenos = df_sorted[df_sorted['indice'] == 'Bueno']['Escenario'].tolist()

# 6. Crear objeto JSON
output = {
    "escenarios_buenos": buenos
}

# 7. Imprimir o guardar JSON
print(json.dumps(output, indent=2))
# Opcional: guardarlo en un archivo
# with open('escenarios_buenos.json', 'w') as f:
#     json.dump(output, f, indent=2)
```
